2. Data Pre-processing

Normalize Data + Splitting of Data

```r
A = data
normalize <- function(x) {
  return ((x - min(x)) / (max(x) - min(x)))
}

#Feature Scaling
A <- A[,-1]
continuous = c(1, 5, 12:23)
continuous_col = A[,continuous]
categorical_col = A[, c(2,3,4,6:11)]
normalize_cont <- as.data.frame(lapply(continuous_col, normalize))
n_sex <- to.dummy(A$SEX,"SEX")
n_education <- to.dummy(A$EDUCATION,"EDUCATION")
n_marriage <- to.dummy(A$MARRIAGE,"MARRIAGE")
n_pay0 <- to.dummy(A$PAY_0,"PAY_O_")
n_pay2 <- to.dummy(A$PAY_2,"PAY_2_")
n_pay3 <- to.dummy(A$PAY_3,"PAY_3_")
n_pay4 <- to.dummy(A$PAY_4,"PAY_4_")
n_pay5 <- to.dummy(A$PAY_5,"PAY_5_")
n_pay6 <- to.dummy(A$PAY_6,"PAY_6_")
normalize_data <- data.frame(normalize_cont,n_sex,n_education,n_marriage,n_pay0,n_pay2,n_pay3,n_pay4,n_pay5,n_pay6,A$`default payment next month`)

set.seed(123)
n = length(normalize_data$LIMIT_BAL)
index <- 1:nrow(normalize_data)
testindex <- sample(index,trunc(2*n)/3)
test.data <- normalize_data[testindex,]
train.data <- normalize_data[-testindex,]
train.data$A..default.payment.next.month. <- factor(train.data$A..default.payment.next.month.) 

ub.train.data <- train.data
ub.test.data <- test.data
```

4.1 SVM

4.1.1 Original Train Dataset

```r
d<-data.frame(matrix(vector(),ncol=7))
colnames(d)<-c("Kernel Type","Type of training data", "Accuracy", "AUC", "Precision" , "Recall", "F1 Score")
ktypes = list("linear", "polynomial", "radial", "sigmoid")

for (i in ktypes) {
classifier = svm(formula = A..default.payment.next.month. ~ ., 
                 data = ub.train.data, 
                 type = 'C-classification', 
                 kernel = i, scale = FALSE) 
testdata <- ub.test.data[-length(ub.test.data)]
y_pred = predict(classifier, newdata=ub.test.data)
cm = table(ub.test.data[,length(ub.test.data)],y_pred)
y <- as.factor(ub.test.data[,length(ub.test.data)])
precision <- diag(cm)[2] / apply(cm, 2, sum)[2]
recall <- diag(cm)[2] / apply(cm, 1, sum)[2]
n = sum(cm) # number of instances
diag = diag(cm) # number of correctly classified instances per class 
accuracy = sum(diag) / n 
kerneltype = i
traindatatype = "Full training data - Unbalanced"
f1 = 2 * precision * recall / (precision + recall) 
library(ROCR)
pred <- prediction(as.numeric(y_pred), as.numeric(ub.test.data[,length(ub.test.data)]))
auc = performance(pred,"auc")
auc = unlist(slot(auc,"y.values"))
b <- data.frame(kerneltype, traindatatype, accuracy,auc, precision, recall, f1)
d <- rbind(d , b)
}
cm
```

```r
#Boosting on unbalanced data
library(xgboost)
train.label = as.integer(ub.train.data$A..default.payment.next.month.) - 1
test.label = as.integer(test.data$A..default.payment.next.month.) - 1
xgb.train = xgb.DMatrix(data=as.matrix(ub.train.data[,-which(names(ub.train.data) == "A..default.payment.next.month.")]),label=train.label)
xgb.test = xgb.DMatrix(data=as.matrix(test.data[,-which(names(test.data) == "A..default.payment.next.month.")]),label=test.label)
num_class = length(levels(test.data$A..default.payment.next.month.))
params = list(
  booster="gbtree",
  eta=0.001,
  max_depth=5,
  gamma=3,
  subsample=0.75,
  colsample_bytree=1,
  objective="multi:softprob",
  eval_metric="mlogloss",
  num_class=num_class
)
#Training the model
xgb.fit=xgb.train(
  params=params,
  data=xgb.train,
  nrounds=10000,
  nthreads=1,
  early_stopping_rounds=10,
  watchlist=list(val1=xgb.train,val2=xgb.test),
  verbose=0
)
# Predict outcomes with the test data
xgb.pred = predict(xgb.fit,as.matrix(test.data[,-which(names(test.data) == "A..default.payment.next.month.")]),reshape=T)
xgb.pred = as.data.frame(xgb.pred)
colnames(xgb.pred) = levels(test.data$A..default.payment.next.month.)
#Get predicted values
xgb.pred$prediction = apply(xgb.pred,1,function(x) colnames(xgb.pred)[which.max(x)])
confusionMatrix(factor(xgb.pred$prediction), factor(test.data$A..default.payment.next.month.),positive = "Yes",mode = "everything")

#ROC
 
newPred = prediction(xgb.pred[,"Yes"], test.data$A..default.payment.next.month.)
roc.perf = performance(newPred, measure = "tpr", x.measure = "fpr")
plot(roc.perf)
roc.perf = performance(newPred, measure = "auc")
roc.perf@y.values 
```
